```python
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Part 1: Iris dataset

```python
fileURL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
data = pd.read_csv(fileURL, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'],header=None)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
n = data.shape[1]
for i in range(n-1):
    plt.hist(data.iloc[:, i])
    colname=list(data)[i]
    plt.xlabel("%s" % colname)
    plt.show()
```

```python
data.boxplot(grid=False, return_type='axes')
```

# Part 3: Conceptual questions

### The histograms for sepal length and sepal width are similar to bell shape, while the histograms for petal length and petal width are similar to bimodal distribution. The distribution of petal lengths can be segmented into two parts at roughly 2.5.

### 'Sepal Length' and 'Petal Width' appear to have significantly different medians. Based on the box plots, 'Petal Length' appear to explain the greatest amount of the data.

### Outliers were observed for feature 2 and feature 4. The histograms of these two features are both left-skewed, where most of the instances are having large values for these two features. Therefore, instances with small values for these two features would most likely be considered as outliers.
